```
import numpy as np
# import pplots
from pplots import plot_embedding

```

```
n = 5
emb = np.random.uniform(0, 1, [n, 2])
labels = ['point'+str(i) for i in range(n)]
```

```
plot_embedding(
        emb, 
        labels,
        show_lines=True,
        show_text=True,
        title='My random example',
        # file_name='example.png'
        file_name=None
        )
```
